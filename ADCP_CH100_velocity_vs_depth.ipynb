```python
#block 1: Import libraries useful for the script.

%matplotlib inline
import scipy.io as sio           
import matplotlib.pyplot as plt
import numpy as np
```

```python
#block 2: Import the data file (matlab to python).

data = sio.loadmat('/media/z3525831/Transcend/UNSW/DATA ADCP/CH100_nostructure.mat')
```

```python
#block 3: Import of variables U, V and time

data_U = data['U']
data_V = data['V']
data_t = data['t']
```

```python
#block 4: Convert the ADCP time serie dataset (Matlab to python time serie: yyyy/mm/dd hh/mm/ss)

t = data_t[0,:]   
import datetime as dt
def matlab2datetime(matlab_datenum):
    day = dt.datetime.fromordinal(int(matlab_datenum))                       #Conversion of Matlab time data to  
    dayfrac = dt.timedelta(days=matlab_datenum%1) - dt.timedelta(days = 366) #classic python time data.
    return day + dayfrac

# convert Matlab variable "t" into list of python datetime objects.
py_t = [matlab2datetime(tval) for tval in t]
```

```python
#block 5: Creation of variable 'U' and 'V' for three depth values (13m,49m,89m).
# [19,:]= 13m ; [10,:]= 49m  ;  [0,:] = 89m

U13 = data_U[19,:]
V13 = data_V[19,:]       
U49 = data_U[10,:]       
V49 = data_V[10,:]
U89 = data_U[0,:]
V89 = data_V[0,:]
```

```python
#block 6: Final plot of the U and V variables for the depth 13m,49m,89m.

plt.rcParams['figure.figsize'] = (12,8)
t = data_t[0,:]               
plt.plot(py_t,U13,"r", label="U13") 
plt.plot(py_t,V13, label="V13")       
plt.xlabel("Time")                                                      
plt.ylabel("velocity (m.s-1)")    
plt.title("CH100 velocity data vs time for depth 13m")      
plt.grid(True)                        
plt.legend()                          
plt.show()

plt.rcParams['figure.figsize'] = (12,8)                              
t = data_t[0,:]               
plt.plot(py_t,U49,"r", label="U49")
plt.plot(py_t,V49, label="V49")
plt.xlabel("Time")             
plt.ylabel("velocity (m.s-1)")
plt.title("CH100 velocity data vs time for depth 49m")
plt.grid(True)                       
plt.legend() 
plt.show()


plt.rcParams['figure.figsize'] = (12,8)                            
t = data_t[0,:]               
plt.plot(py_t,U89,"r", label="U89")
plt.plot(py_t,V89, label="V89")
plt.xlabel("Time")             
plt.ylabel("velocity (m.s-1)")
plt.title("CH100 velocity data vs time for depth 89m")
plt.grid(True)                       
plt.legend() 
plt.show()
```
